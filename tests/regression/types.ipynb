# Typed functions, signatures, structs, and enums

```smlnj
fun simple(x: int) : int =
    x + 1
```

```smlnj
signature TestSig =
sig
    val x : int
    val y : int
    val add : int -> int -> int
end
```

```smlnj
structure TestStruct : TestSig = 
struct
    val x = 5
    val y = 10
    val add = fn (a: int) => fn (b: int) => a + b
end
```

```smlnj
signature ORDERED =
sig
    type T
    
    val eq : T * T -> bool
    val lt : T * T -> bool
    val leq : T * T -> bool
end
```

```smlnj
structure IntOrdered : ORDERED =
struct
    type T = int

    fun eq (x: int, y: int) = x = y
    fun lt (x: int, y: int) = x < y
    fun leq (x: int, y: int) = x <= y
end
```
